### 3. For each channel, determine the total number of views

Since there are several videos that have trended multiple times and in different countries, the logic was to keep only the row with the last 'trending_date', eliminating any duplicates.
The total number of views is an objective metric, so the view counts from different countries but taken in the same day are equal.

However there are some problems: there are some videos with no video_id, instead there is '#NAME?' or '#VALUE!' in the column video_id. so in the parameter subset of the function drop_duplicates, that tels on which columns controll the possible duplicates, I select 'video_id', 'channel_title' and 'publish_time'.

```python
df.sort_values(['video_id', 'trending_date']).drop_duplicates(subset = ['video_id', 'publish_time', 'channel_title'], keep = 'last').groupby('channel_title').sum()[['views']]
```

### 4. Save all rows with disabled comments and disabled ratings, or that have video_error_or_removed in a new dataframe called excluded, and remove those rows from the original dataframe.

### 5. Add a like_ratio column storing the ratio between the number of likes and of dislikes

Using np.where to calculate 'like_ratio' conditionally:
- Compute likes / dislikes only when both likes and dislikes are greater than 0.
- If either is 0, assign None to avoid division by zero or meaningless ratios.

```python
df_new['like_ratio'] = np.where(
    (df_new['likes'] > 0) & (df_new['dislikes'] > 0),
    df_new['likes'] / df_new['dislikes'],
    None
)
df_new[df_new['like_ratio'].isnull()]
```

### 6. Cluster the publish time into 10-minute intervals (e.g. from 02:20 to 02:30)

```python
df_new['publish_time'] = pd.to_datetime(df_new['publish_time'])
df_new['publish_time'].dtypes
```

```python
# floor the time to nearest 10-minutes starting point (round down)
start_times = df_new['publish_time'].dt.floor('10min')
end_times = start_times + pd.Timedelta(minutes=10)

# extract only hour and minutes for labels
df_new['cluster_hour'] = (
    "from " + start_times.dt.strftime('%H:%M') + 
    " to " + end_times.dt.strftime('%H:%M')
)
```

### 7. For each interval, determine the number of videos, average number of likes and of dislikes.

The logic is the same of 3° point, keeping only the last 'trending_date' and deleting others.

```python
df_unique = df_new.sort_values(['video_id', 'trending_date']).drop_duplicates(
    subset=['video_id', 'publish_time', 'channel_title'], 
    keep='last'
).copy()
```

```python
# group by created clusters and aggregate metrics
df_unique.groupby('cluster_hour').agg(
    n_videos=('video_id', 'count'),
    avg_likes=('likes', 'mean'),
    avg_dislikes=('dislikes', 'mean')
).round(2)
```
