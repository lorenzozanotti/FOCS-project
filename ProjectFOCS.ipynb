### 5. Add a like_ratio column storing the ratio between the number of likes and of dislikes

Using np.where to calculate 'like_ratio' conditionally:
- Compute likes / dislikes only when both likes and dislikes are greater than 0.
- If either is 0, assign None to avoid division by zero or meaningless ratios.

```python
df_new['like_ratio'] = np.where(
    (df_new['likes'] > 0) & (df_new['dislikes'] > 0),
    df_new['likes'] / df_new['dislikes'],
    None
)
df_new[df_new['like_ratio'].isnull()]
```

### 7. For each interval, determine the number of videos, average number of likes and of dislikes.

```python
df_new.set_index('publish_time').resample('10min').agg(
    n_videos=('video_id', 'count'),
    avg_likes=('likes', 'mean'),
    avg_dislikes=('dislikes', 'mean')
).dropna(subset=['avg_likes', 'avg_dislikes']).round({'avg_likes': 2, 'avg_dislikes': 2})
```

### 14. Read all json files with the video categories

In the following code block, we read JSON files containing YouTube video categories for different countries. Each JSON file has a nested structure in which the relevant information is stored under the key "items".

We use the function pd.read_json() to import each file into a Pandas object, and then apply pd.json_normalize() to flatten the nested structure, transforming the list of categories into a tabular DataFrame, which is more suitable for data analysis and manipulation.

```python
CA_json =pd.read_json("C:\\Users\\loren\\OneDrive\\Desktop\\Data Science\\Intro to Computer Science\\trendingYT\\CA_category_id.json")
CA_json =pd.json_normalize(CA_json["items"])
DE_json =pd.read_json("C:\\Users\\loren\\OneDrive\\Desktop\\Data Science\\Intro to Computer Science\\trendingYT\\DE_category_id.json")
DE_json =pd.json_normalize(DE_json["items"])
FR_json =pd.read_json("C:\\Users\\loren\\OneDrive\\Desktop\\Data Science\\Intro to Computer Science\\trendingYT\\FR_category_id.json")
FR_json =pd.json_normalize(FR_json["items"])
GB_json =pd.read_json("C:\\Users\\loren\\OneDrive\\Desktop\\Data Science\\Intro to Computer Science\\trendingYT\\GB_category_id.json")
GB_json =pd.json_normalize(GB_json["items"])
IN_json =pd.read_json("C:\\Users\\loren\\OneDrive\\Desktop\\Data Science\\Intro to Computer Science\\trendingYT\\IN_category_id.json")
IN_json =pd.json_normalize(IN_json["items"])
JP_json =pd.read_json("C:\\Users\\loren\\OneDrive\\Desktop\\Data Science\\Intro to Computer Science\\trendingYT\\JP_category_id.json")
JP_json =pd.json_normalize(JP_json["items"])
KR_json =pd.read_json("C:\\Users\\loren\\OneDrive\\Desktop\\Data Science\\Intro to Computer Science\\trendingYT\\KR_category_id.json")
KR_json =pd.json_normalize(KR_json["items"])
MX_json =pd.read_json("C:\\Users\\loren\\OneDrive\\Desktop\\Data Science\\Intro to Computer Science\\trendingYT\\MX_category_id.json")
MX_json =pd.json_normalize(MX_json["items"])
RU_json =pd.read_json("C:\\Users\\loren\\OneDrive\\Desktop\\Data Science\\Intro to Computer Science\\trendingYT\\RU_category_id.json")
RU_json =pd.json_normalize(RU_json["items"])
US_json =pd.read_json("C:\\Users\\loren\\OneDrive\\Desktop\\Data Science\\Intro to Computer Science\\trendingYT\\US_category_id.json")
US_json =pd.json_normalize(US_json["items"])
```
